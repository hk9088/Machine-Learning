```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np


# Loading the database
iris = load_iris()
data = np.array(iris['data'])
targets = np.array(iris['target'])


cd = {0:'r',1:'b',2:"g"}    # Color dictionary
cols = np.array([cd[target] for target in targets])

 
fig = plt.figure(figsize=(8,8))   # Creating a figure with size(8, 8)

# ploting data
plt.subplot(4,3,1).scatter(data[:,0], data[:,1], c=cols, edgecolors='black')
plt.subplot(4,3,2).scatter(data[:,0], data[:,2], c=cols, edgecolors='black')
plt.subplot(4,3,3).scatter(data[:,0], data[:,3], c=cols, edgecolors='black')
plt.subplot(4,3,4).scatter(data[:,1], data[:,0], c=cols, edgecolors='black')
plt.subplot(4,3,5).scatter(data[:,1], data[:,2], c=cols, edgecolors='black')
plt.subplot(4,3,6).scatter(data[:,1], data[:,3], c=cols, edgecolors='black')
plt.subplot(4,3,7).scatter(data[:,2], data[:,0], c=cols, edgecolors='black')
plt.subplot(4,3,8).scatter(data[:,2], data[:,1], c=cols, edgecolors='black')
plt.subplot(4,3,9).scatter(data[:,2], data[:,3], c=cols, edgecolors='black')
plt.subplot(4,3,10).scatter(data[:,3], data[:,0], c=cols, edgecolors='black')
plt.subplot(4,3,11).scatter(data[:,3], data[:,1], c=cols, edgecolors='black')
plt.subplot(4,3,12).scatter(data[:,3], data[:,2], c=cols, edgecolors='black')

plt.show()  # Show plotted data
```
