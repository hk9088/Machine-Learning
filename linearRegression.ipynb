```python
def gradient(x,y,theta):
    
    alpha = 0.00001
    iteration = 2000
    #gradient descend algorithm
    J_history = np.zeros([iteration, 1]);
    for iter in range(0,2000):
        error = (x @ theta) -y
        temp0 = theta[0] - ((alpha/m) * np.sum(error*x[:,0]))
        temp1 = theta[1] - ((alpha/m) * np.sum(error*x[:,1]))
        theta = np.array([temp0,temp1]).reshape(2,1)
        J_history[iter] = (1 / (2*m) ) * (np.sum(((x @ theta)-y)**2))   
        #compute J value for each iteration 
    return theta, J_history
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_diabetes

# Linear Regression function
def LinearRegression(x,y):
  # this funtion does gradient descent
  # return bias and gradient descent
  
  m = ((np.mean(x)*np.mean(y)) - np.mean(x * y)) / (np.mean(x)**2 - np.mean(x**2))
  b = np.mean(y) - ( m * np.mean(x))
  
  return [m,b]

d = load_diabetes()             # loading the data
d_X = d.data[:, np.newaxis, 2]

# Only using last 20 data
dx_train = d_X[:-20]
dy_train = d.target[:-20]
dx_test = d_X[-20:]
dy_test = d.target[-20:]

m, b = LinearRegression(dx_train.squeeze(), dy_train)

# line equation
y_predic = m*d_X + b

plt.scatter(dx_train,dy_train,c='r')    # scattetring training data
plt.scatter(dx_test,dy_test,c='g')      # scattetring test data
plt.plot(d_X,y_predic,c='b')            # ploting best fit line
# add legend
plt.legend(['Prediction','Train', 'Test'])

plt.show()  
```
